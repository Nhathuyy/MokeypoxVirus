```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
dcwcc = pd.read_csv('https://raw.githubusercontent.com/Nhathuyy/MokeypoxVirus/main/Daily_Country_Wise_Confirmed_Cases.csv')
mpcw = pd.read_csv('https://raw.githubusercontent.com/Nhathuyy/MokeypoxVirus/main/Monkey_Pox_Cases_Worldwide.csv')
wcdt = pd.read_csv('https://raw.githubusercontent.com/Nhathuyy/MokeypoxVirus/main/Worldwide_Case_Detection_Timeline.csv')
```

```python
mpcw[['Confirmed_Cases','Suspected_Cases']] = mpcw[['Confirmed_Cases','Suspected_Cases']].astype(int)
print_df = mpcw[['Country','Confirmed_Cases','Suspected_Cases']].sort_values('Confirmed_Cases',ascending=False)
top_countries = print_df.Country[:6]
```

```python
print_df.style.background_gradient(cmap="Reds")
```

```python
dates = list(dcwcc.columns)
dates.remove('Country')
new_cases = pd.DataFrame(dcwcc[dates].sum(),columns=['number_of_cases'])
new_cases = new_cases.reset_index(level=0)
new_cases = new_cases.rename(columns={'index':'Date'})
```

```python
fig = px.line(new_cases, x=new_cases.Date, y='number_of_cases',markers=True)
fig.show()
```

```python
#Total monkey pox cases linear scale
new_cases['total_cases'] = np.nan

for index,num in enumerate(range(len(new_cases))):
    new_cases.loc[index,'total_cases'] = sum(new_cases.number_of_cases[:index].values)
```

```python
fig = px.line(new_cases, x=new_cases.Date, y='total_cases',markers=True)
fig.show()
```

```python
top_country_outbreak = dcwcc[dcwcc['Country'].isin(top_countries)].reset_index(drop=True)
top_country_outbreak = top_country_outbreak.T.reset_index(level=0)
top_country_outbreak.columns = top_country_outbreak.loc[0]
top_country_outbreak = top_country_outbreak[1:].reset_index(drop=True).rename(columns={'Country':'Date'})
```

# Top 6 Countries Daily Reported Cases

```python
top_6_outbreak = pd.DataFrame()

for c in top_countries:
    temp_df = top_country_outbreak[['Date',c]]
    temp_df = temp_df.rename(columns={c:'daily number of cases'})
    temp_df['Country'] = c
    top_6_outbreak = top_6_outbreak.append(temp_df)
    
top_6_outbreak.reset_index(drop=True,inplace=True)
```

```python
fig = px.line(top_6_outbreak, x="Date", y="daily number of cases", color='Country')
fig.show()
```

# Top 6 Countries Total Cases : Linear scale

```python
top_6_outbreak['total_cases'] = np.nan
country = top_countries[0]
count = top_6_outbreak['daily number of cases'][0]
for ind in range(len(top_6_outbreak)):
    if(top_6_outbreak["Country"][ind]!=country):
        count = top_6_outbreak['daily number of cases'][ind]
        country = top_6_outbreak['Country'][ind]
        
    count = count+top_6_outbreak['daily number of cases'][ind]
    top_6_outbreak.loc[ind,'total_cases'] = count
```

```python
fig = px.line(top_6_outbreak, x="Date", y="total_cases", color='Country')
fig.show()
```
